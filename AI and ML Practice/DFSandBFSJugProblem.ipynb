```python
## I assisted Ashar Nadeem and gave him an idea of how to develop this algorithm and checking these states 
def childrenOfState(currentState, endStateTotal, totalCap):
    jug1 = currentState[0]
    jug2 = currentState[1]
    jug3 = currentState[2]

    jugsTotal = jug1 + jug2 + jug3
    
    q = []
    
    jug1Capacity = totalCap[0]
    jug2Capacity = totalCap[1]
    jug3Capacity = totalCap[2]
    
    if(jugsTotal >= endStateTotal):
        
        emptyJug1 = (0,jug2,jug3)
        emptyJug2 = (jug1,0,jug3)
        emptyJug3 = (jug1,jug2,0)

        if(jug1 != 0 and emptyJug1 != currentState):
            q.append(emptyJug1)

        if(jug2 != 0 and emptyJug2 != currentState):
            q.append(emptyJug2)

        if(jug3 != 0 and emptyJug3 != currentState):
            q.append(emptyJug3)


        #------------------------------------------------------------
        #dumping jug1 into jug2 if jug2 is empty or has capacity > jug2+jug1 (sizes)
        if(jug1+jug2 <= jug2Capacity):
            
            state = (0,jug1+jug2,jug3)
            if(state != currentState):
                q.append(state)

        elif(jug2 < jug2Capacity and jug1 > (jug2Capacity - jug2)):
            state = (jug1-(jug2Capacity-jug2),jug2Capacity,jug3)
            if(state != currentState):
                q.append(state)    

        #------------------------------------------------------------
        #dumping jug2 into jug1 if jug1 is empty or has capacity > jug2+jug1 (sizes)

        if(jug1+jug2 <= jug1Capacity):
            
            state = (jug1+jug2,0,jug3)
            if(state != currentState):
                q.append(state)

        elif(jug1 < jug1Capacity and jug2 > (jug1Capacity - jug1)):
            state = (jug1Capacity,jug2-(jug1Capacity-jug1),jug3)
            if(state != currentState):
                q.append(state)


        #-------------------------------------------------------------
        #dumping jug1 into jug3 if jug3 is empty or has capacity > jug3+jug1 (sizes)

        if(jug1+jug3 <= jug3Capacity):
            
            state = (0,jug2,jug1+jug3)
            if(state != currentState):
                q.append(state)

        elif(jug3 < jug3Capacity and jug1 > (jug3Capacity-jug3)):
            state = (jug1-(jug3Capacity-jug3),jug2,jug3Capacity)
            if(state != currentState):
                q.append(state)

        #-------------------------------------------------------------
        #dumping jug2 into jug3 if jug3 is empty or has capacity > jug3+jug2 (sizes)

        if(jug2+jug3 <= jug3Capacity):
            
            state = (jug1,0,jug2+jug3)
            if(state != currentState):
                q.append(state)

        elif(jug3 < jug3Capacity and jug2 > (jug3Capacity-jug3)):
            state = (jug1,jug2-(jug3Capacity-jug3),jug3Capacity)
            if(state != currentState):
                q.append(state)

        #-------------------------------------------------------------
        #dumping jug3 into jug2 if jug2 is empty or has capacity > jug3+jug2 (sizes)

        if(jug2+jug3 <= jug3Capacity):
            state = (jug1,jug2+jug3,0)
            if(state != currentState):
                q.append(state) 

        elif(jug2 < jug2Capacity and jug3 > (jug2Capacity-jug2)):
            state = (jug1,jug2Capacity,jug3-(jug2Capacity-jug2))
            if(state != currentState):
                q.append(state)

        #-------------------------------------------------------------
        #dumping jug3 into jug1 if jug1 is empty or has capacity > jug3+jug1 (sizes)

        if(jug1+jug3 <= jug2Capacity):
            state = (jug1+jug3,jug2,0)
            if(state != currentState):
                q.append(state)

        elif(jug1 < jug1Capacity and jug1 > (jug3Capacity-jug3)):
            state = (jug1-(jug3Capacity-jug3),jug2,jug3Capacity)
            if(state != currentState):
                q.append(state)
            
        return q
        
```

```python
def jugs(jugsize1, jugsize2, jugsize3, endstate1, endstate2, endstate3, algorithm):
    
    jug1Capacity = jugsize1
    jug2Capacity = jugsize2
    jug3Capacity = jugsize3
    
    endStateTotal = endstate1 + endstate2 + endstate3
    totalCap = (jug1Capacity,jug2Capacity,jug3Capacity)
    
    # array that holds the path and nodes visited
    visitedStates = []
    # should hold the correct path after triversal 
    correctPath = []
    # holds the list and states together
    allStatePaths = []
    
    parentPathDict = {}    
    q = []

    # initial conditions
    q.append((jugsize1,jugsize2,jugsize3))

    canBeSolved = False    
    nodeCounter = 0
    
    # checking the algorithm 
    
    if(algorithm == "bfs"):
        while(len(q) > 0):
            nodeCounter += 1
            currentState = q[0]    

            if(currentState[0] == (endstate1) and currentState[1] == endstate2 and currentState[2] == endstate3):
                visitedStates.append(currentState)
                break

            jugSizeTotal = currentState[0] + currentState[1] + currentState[2]

            # if current state is the end state
            # print the parents path plus add this currentstate 
            
            # I am initializing this variable to see if the node is in the visited State
            # if it is the loop ends, stops searching visited states and c
            continuePast = 1

            q.pop(0)

            # if the state is already visisted
            for i in range(0,len(visitedStates)):
                if(visitedStates[i] == currentState):
                    continuePast = 0
                    break

            # if we arent meeting the restraints of the endstates 4+4+1 > 1+1+1 but 1+0+1 < 1+1+1 impossible        
            if((jugSizeTotal) < endStateTotal or continuePast == 0):
                continue

            visitedStates.append(currentState)
            
            for i in range(0,len(childrenOfState(currentState,endStateTotal,totalCap))):
                q.append(childrenOfState(currentState,endStateTotal,totalCap)[i])
                parentPathDict[(childrenOfState(currentState,endStateTotal,totalCap)[i],nodeCounter)] = (currentState,nodeCounter)

        
        searchState = (endstate1,endstate2,endstate3)
        searchState1 = searchState
        
        # strange way to output the dictionary but basically i set the parent to the child
        # and keep the cycle going but it finds the first one and contiunes adding the child to the 
        # correct path until it hits the final parent 
        for i in parentPathDict:

            for j in range(0,nodeCounter):
                if(searchState not in correctPath):  
                    correctPath.append(searchState)
                try:
                    searchState1 = parentPathDict[searchState,j] 
                    break
                except:
                    continue
            searchState = searchState1[0]

    jug1Capacity = jugsize1
    jug2Capacity = jugsize2
    jug3Capacity = jugsize3
    
   
    endstateTotal = endstate1 + endstate2 + endstate3
    
    # need to keep track of the visited states
    visitedStates = []
    
    ## PreImplemented DFS so i just attached it with an if statement and it works 
    if(algorithm == "dfs"):
        
        def fillTree(currentState):
            
            jug1 = currentState[0]
            jug2 = currentState[1]
            jug3 = currentState[2]

            if(jug1 == endstate1 and jug2 == endstate2 and jug3 == endstate3):
                correctPath.append(currentState)
                return True

            if((currentState) in visitedStates):
                return False

            visitedStates.append((jug1,jug2,jug3))

            jugsTotal = jug1 + jug2 + jug3

            # JUG 1
            if(jug1 > 0):
                
                #dumping jug1 if the total is greater than the total of endStates 4+4+1 > 1+1+1
                if(jugsTotal > endstateTotal):
                    if(fillTree((0,jug2,jug3))):
                        correctPath.append(currentState)
                        return True
                #dumping jug1 into jug2 if jug2 is empty or has capacity > jug2+jug1 (sizes)
                if(jug1+jug2 <= jug2Capacity):
                    if(fillTree((0,jug1+jug2,jug3))):
                        correctPath.append(currentState)
                        return True
                else:
                    if(fillTree((jug1-(jug2Capacity-jug2),jug2Capacity,jug3))):
                        correctPath.append(currentState)
                        return True

                #dumping jug1 into jug3 if jug2 is empty or has capacity > jug3+jug1 (sizes)
                if(jug1+jug3 <= jug3Capacity):
                    if(fillTree((0,jug2,jug1+jug3))):
                        correctPath.append(currentState)
                        return True
                else:
                    if(fillTree((jug1-(jug3Capacity-jug3),jug2,jug3Capacity))):
                        correctPath.append(currentState)
                        return True

            # JUG 2
            if(jug2 > 0):
                #dumping jug1 if the total is greater than the total of endStates 4+4+1 > 1+1+1
                if(jugsTotal > endstateTotal):
                    if(fillTree((jug1,0,jug3))):
                        correctPath.append(currentState)
                        return True
                #dumping jug2 into jug1 if jug1 is empty or has capacity > jug2+jug1 (sizes)
                if(jug1+jug2 <= jug1Capacity):
                    if(fillTree((jug1+jug2,0,jug3))):
                        correctPath.append(currentState)
                        return True
                #dumping jug2 into jug3 if jug3 is empty or has capacity > jug3+jug2 (sizes)
                if(jug2+jug3 <= jug3Capacity):
                    if(fillTree((jug1,0,jug2+jug3))):
                        correctPath.append(currentState)
                        return True  

            # JUG 3 
            if(jug3 > 0):
                #dumping jug3 if the total is greater than the total of endStates 4+4+1 > 1+1+1
                if(jugsTotal > endstateTotal):
                    if(fillTree((jug1,jug2,0))):
                        correctPath.append(currentState)
                        return True
                #dumping jug3 into jug1 if jug1 is empty or has capacity > jug3+jug1 (sizes)
                if(jug1+jug3 <= jug2Capacity):
                    if(fillTree((jug1+jug3,jug2,0))):
                        correctPath.append(currentState)
                        return True
                #dumping jug1 into jug3 if jug2 is empty or has capacity > jug3+jug1 (sizes)
                if(jug2+jug3 <= jug3Capacity):
                    if(fillTree((jug1,jug2+jug3,0))):
                        correctPath.append(currentState)
                        return True  

            return False
        fillTree((jugsize1,jugsize2,jugsize3))
    
    correctPath.reverse()
    if(len(correctPath) > 2):
        print("Solution Path Found: ")
        for i in correctPath:
            print(i)
    else:
        print("No Solution Found")
    
    
```

```python
## https://medium.com/cantors-paradise/dijkstras-shortest-path-algorithm-in-python-d955744c7064 
## ^ website above helped me understand init of classes and making this class

class State:
    def __init__(self, node):
        self.currentX = node[0]
        self.currentY = node[1]
        
        self.weight = None
        self.valueOfWeight = None
        
        self.node = node
        self.parent = None
        self.currentPath = [node]
    
      
    def neighbors(self):
        neighbors = []
        
        ## making sure the values are in bounds 10x10 grid == 0-9
        # checking above current
        if(self.currentY < 9):
            neighbors.append([self.currentX,self.currentY + 1])
        # checking below current
        if(self.currentY > 0):
            neighbors.append([self.currentX,self.currentY - 1])
        # checking to the right
        if(self.currentX < 9):
            neighbors.append([self.currentX + 1,self.currentY])
        # checking to the left
        if(self.currentX > 0):
            neighbors.append([self.currentX - 1,self.currentY])
            
        return neighbors
    
    # when looking through the neighbor node you need to maintain clearing the path
    def relax(self):
        self.currentPath = []
        for node in self.parent.currentPath:
            # need to make a copy instead of actually setting equal to
            nodeCopy = node.copy()
            self.currentPath.append(nodeCopy)
        # adding the original point right back to the emtpy list
        self.currentPath.append(self.node)
    
    ## getting the string implimentation of the node 
    def __str__(self):
        if self.weight is None:
            ## if i have not touched the node aka not in 'neighborhood' i print the *
            return str("_")
        else:
            # if i have a weight listed for such node just returns its weight
            return str(self.weight)
```

```python
# implementation taken from wiki page of dijkstra, it had a pretty good psuedo code
import math

def findingPath(mymap,alg):
    unvisitedPath = []
    cellsvisited = []
    
    for y in range(len(mymap)):
        row = []
        for x in range(len(mymap[y])):
            row.append(State([x, y]))
            unvisitedPath.append([x,y])
        cellsvisited.append(row)
    
    startingX = 0
    startingY = 0
    endX = 9
    endY = 9
    
    # always starting from (0,0) index so always start, so that falls into a state visited 
    startingCell = cellsvisited[startingX][startingY]
    startingCell.weight = mymap[startingCell.currentX][startingCell.currentY]
    
    endingCell = cellsvisited[endX][endY]
    allSolutions = []
    
    def heuristic(node):
        # i honestly tried some stuff and am not sure if i can get this working 
        # it doesnt seem to make a difference 
        
        ### Euclidean Distance --> couldnt get sqrt to work
        #h1 = math.sqrt((node.currentX-endX)*2 + (node.currentY-endY)*2)
        
        ### Diagonal Distance
        h2 = max(abs(node.currentX-endX),abs(node.currentY-endY))
        
        ### Manhattan Distance --> seemed to work the best based on my testing
        h3 = abs(node.currentX-endX) + abs(node.currentY-endY)
        
        return h3
    
    # 1 is astar
    if(alg == "astar"):
        startingCell.valueOfWeight = heuristic(startingCell)
    printlist = []
    counter = 0
    while(1):
        counter = counter + 1
        lightestWeight = None
        for i in unvisitedPath:
            currentCell = cellsvisited[i[1]][i[0]]
            if(currentCell.weight is not None and (lightestWeight is None or currentCell.weight < lightestWeight.weight)):
                lightestWeight = currentCell
            if(alg == "astar"):    
                if(currentCell.valueOfWeight is not None and (lightestWeight is None or currentCell.valueOfWeight < lightestWeight.valueOfWeight)):
                    lightestWeight = currentCell
        #------------------------------------------------------#
        for nextDoor in lightestWeight.neighbors():
            if(nextDoor in unvisitedPath):
                nextDoorCell = cellsvisited[nextDoor[1]][nextDoor[0]]
                
                smallestWeight = lightestWeight.weight + mymap[nextDoor[1]][nextDoor[0]]
                if(nextDoorCell.weight is None or nextDoorCell.weight > smallestWeight):
                    nextDoorCell.weight = smallestWeight
                    if(alg == "astar"):
                        nextDoorCell.valueOfWeight = heuristic(nextDoorCell)
                    nextDoorCell.parent = lightestWeight
                    nextDoorCell.relax()
                    
        unvisitedPath.remove(lightestWeight.node)
        allSolutions.append(cellsvisited)
        
        # Prof Don said to cut if off around 20 prints
        if(counter == 1):
            print("Starting to print the path: ")
            print()
        printlist = []
        for row in cellsvisited:
            printrow = []
            for i in row:
                printrow.append(str(i))
            printlist.append(printrow)
            
        for row in printlist:
            print("   ", ("{:>4}" * len(cellsvisited)).format(*row))
        print()
           
        if(lightestWeight == endingCell):
            break
       
    totalNodesUsed = len(mymap[0]) * len(mymap) - len(unvisitedPath)
    return (endingCell.currentPath, endingCell.weight,totalNodesUsed,allSolutions) 
```
